Numpy, Pandas, Matplotlib, and Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the file
df_train = pd.read_csv('house-prices-advanced-regression-techniques/train.csv')
```

```python
#find unique neighbourhood values
unique_neighbour = df_train['Neighborhood'].unique()
print(unique_neighbour)
```

```python
unique_neighbour[0:25]
```

```python
#find the median prices for all neighbourhood
neighbour_median = []
for item in unique_neighbour:
    neighbour_prices = df_train.loc[df_train['Neighborhood'] == item,'SalePrice']
    neighbour_median.append(np.median(neighbour_prices))
neighbour_median
```

```python
#plot bar chart
plt.figure(figsize=(12,7))
plt.bar(unique_neighbour,neighbour_median,color=['blue'])
plt.xlabel('Neighborhood')
ax = plt.xticks(rotation=45)
plt.ylabel('Median price')
plt.title("Median house price of each neighborhood")
plt.show()
```

```python
#Relationship with several numerical variable

#scatter plot Above grade (ground) living area square feet (grlivarea)/saleprice
var = 'GrLivArea'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,900000));
```

```python
#Relationship with categorical feature

#box plot Rates the overall material and finish of the house (OverallQual)/saleprice
var = 'OverallQual'
data = pd.concat([df_train['SalePrice'], df_train[var]], axis=1)
f, ax = plt.subplots(figsize=(10, 8))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=900000);
```
